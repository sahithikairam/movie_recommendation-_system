```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj): #covertion of string of lists into list 
        L.append(i['name'])
    return L 
```

```python
#given genre is string of list so we are converting that into list
#import ast
#ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):  #covertion of string of lists into list 
        if counter !=3:
            L.append(i['name'])
            counter+=1
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
import ast

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):  # conversion of string of lists into list 
        if i['job'] == "Director":
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x if isinstance(x, list) else x.split())
```

```python
# now we should merge colulms for tags,so before that we need to remove white spaces of every string because it may use it as different word 
#ex "Sam Worthington" to "SamWorthington"
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))  #it converts tags into strings
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk  #nlp library
```

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

def stem(text):
    # Stem the text, splitting by space and applying stem to each word
    return " ".join([ps.stem(word) for word in text.split()])

# Apply stemming to each row in the 'tags' column
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
for word in cv.get_feature_names_out():
    print(word)    # it prints all the words we have filtered
```

```python
#we need caluculate distance between vectors of movies with angle(cosine distance) not line distance (euclidian distance)
# use sklean cosine function 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
#cosine_similarity(vectors).shape
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6] #doing distances  reverse with intacted with movies that is why we are using eemerate function
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)

    
```

```python
recommend('Avatar')
```
